```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_runtime_distribution(J, K, C):
    # Construct the file path
    file_path = f'incomplete_global_runtime_J{J}_K{K}_C{C}.csv'

    # Load the CSV file
    df = pd.read_csv(file_path, header=None)

    # Extract the run time column (assuming it's the second column)
    run_times = df[0]

    # Create a histogram
    plt.hist(run_times, bins=30, edgecolor='black')

    # Set the title and labels
    plt.title(f'Distribution of Run Times (J={J}, K={K}, C={C})')
    plt.xlabel('Run Time')
    plt.ylabel('Frequency')

    # Display the plot
    plt.show()
```

```python
# Call the function for different values of J, K, and C
plot_runtime_distribution(J=5, K=4, C=3)
plot_runtime_distribution(J=6, K=4, C=3)
plot_runtime_distribution(J=7, K=4, C=3)
```
