```python
import pandas as pd
import csv
import ast
import numpy as np
import matplotlib.pyplot as plt

def aggregate_data_to_dataframe(filename):
    data = []
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # Skip the header
        for row in reader:
            J = int(row[0])
            K = int(row[1])
            seed = int(row[2])
            
            # Remove surrounding quotes and parse as Python literal
            Q =row[3]
            
            # Convert the string of booleans into a list of booleans
            identifiable_statuses = [s == 'True' for s in row[4][1:-1].split(", ")]
            
            # Convert the string of floats into a list of floats
            elapsed_times = [float(e) for e in row[5][1:-1].split(", ")]
            
            data.append([J, K, seed, Q, identifiable_statuses, elapsed_times])

    df = pd.DataFrame(data, columns=['J', 'K', 'seed', 'Q', 'identifiable_statuses', 'elapsed_times'])

    return df
```

```python
df = aggregate_data_to_dataframe('exprQ_J10_K6.csv')
df
```

```python
# Create the list of all combinations of kappa and check_level
kappas = [1, 2, 3]
check_levels = [3, 4, 5, 6]
column_names = [f"$\kappa${kappa}_$c${check_level}" for kappa in kappas for check_level in check_levels]

# Get the list of elapsed times and create a new DataFrame
elapsed_times_list = df['elapsed_times'].tolist()
elapsed_times_df = pd.DataFrame(elapsed_times_list, columns=column_names)

elapsed_times_df
```

```python
# Define the colors corresponding to different kappas
colors = ['orange', 'c', 'brown']  # Modify this as needed

# Compute the average of each column
avg_elapsed_times = elapsed_times_df.mean()

# Plot histogram with colors
plt.figure(figsize=(10, 6))
for i in range(len(kappas)):
    # Get the average elapsed times for each kappa
    avg_elapsed_times_kappa = avg_elapsed_times[i*len(check_levels):(i+1)*len(check_levels)]
    plt.bar(avg_elapsed_times_kappa.index, avg_elapsed_times_kappa.values, color=colors[i])

plt.xlabel('Setting ($\kappa$ and $c$)')
plt.ylabel('Average Runtime (seconds)')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Average Runtime for Different Settings')

# Add legend
legend_labels = [f'$\kappa$ = {kappa}' for kappa in kappas]
plt.legend(legend_labels)

plt.show()
```

```python
# Convert 'identifiable_statuses' into DataFrame
identifiable_statuses = df['identifiable_statuses'].apply(pd.Series)
identifiable_statuses.columns = column_names

# Convert True/False into 1/0
identifiable_statuses = 1 - identifiable_statuses.astype(int)
identifiable_statuses
```

```python
# Compute the average of each column
avg_identifiable_statuses = identifiable_statuses.mean()

# Define the colors corresponding to different kappas
colors = ['orange', 'c', 'brown']  # Modify this as needed

# Plot histogram with colors
plt.figure(figsize=(10, 6))
for i in range(len(kappas)):
    # Get the average identifiable statuses for each kappa
    avg_identifiable_statuses_kappa = avg_identifiable_statuses[i*len(check_levels):(i+1)*len(check_levels)]
    plt.bar(avg_identifiable_statuses_kappa.index, avg_identifiable_statuses_kappa.values, color=colors[i])

plt.xlabel('Setting ($\kappa$ and $c$)')
plt.ylabel('Average Identifiability (1=True, 0=False)')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Average Identifiability for Different Settings')

# Add legend
legend_labels = [f'$\kappa$ = {kappa}' for kappa in kappas]
plt.legend(legend_labels)

plt.show()
```

```python
import math

def calculate_classes(J, K):
    total_matrices = 2**(J*K)
    total_permutations = math.factorial(K)
    return total_matrices / total_permutations
```
